```python
#Classify Movie Based on Excellent Best Good
def rating(rating):
  if rating>=7.0:
    return "Excellent"
  elif rating>=6.0:
    return "Good"
  else:
    return "Average"
```

```python
df['rating_cat']=df['Rating'].apply(rating)
df.head()
```

```python
#find Unique Genre
list1=[]
for value in df['Genre']:
  list1.append(value.split(','))
```

```python
one_d=[]
for item in list1:
  for item1 in item:
    one_d.append(item1)
```

```python
one_d
```

```python
uni_list=[]
for item in one_d:
  if item not in uni_list:
    uni_list.append(item)
```
