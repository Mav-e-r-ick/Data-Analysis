```python
import matplotlib.pyplot as plt
plt.figure(figsize=(20,6))
sns.heatmap(df.corr(),annot=True)
```

```python
#How many people have heart disease and how many don't
df['target'].value_counts()
```

```python
sns.countplot(x = df['target'])
```

```python
#Find total number of male &female in dataset
df['sex'].value_counts()
```

```python
#Find Gender Distribution according to target value
sns.countplot(x='sex',hue='target',data=df)
plt.xticks([1,0],['Male','Female'])
plt.legend(labels=['No Disease','Disease'])
plt.show()
```

```python
#Check Age Distribution on the dataset
sns.distplot(df['age'],bins=20)
```

```python
#Check Chest pain types
sns.countplot(x=df['cp'])
```

```python
#Show the Chest pain Distribution As per Target Variables
sns.countplot(x='cp', hue='target',data=df)
plt.legend(labels=['No Disease','Disease'])
plt.show()
```

```python
#Fasting blood sugar distribution According to target variable
sns.countplot(x='fbs', hue='target',data=df)
plt.legend(labels=['No Disease','Disease'])
plt.show()
```

```python
#Check Resting Blood Pressure Distribution
df['trestbps'].hist()
```

```python
#Compare esting blood pressure as per sex column
g = sns.FacetGrid(data=df,hue='sex')
g.map(sns.kdeplot,'trestbps',shade=True)
plt.legend(labels=['Male','Female'])
```

```python
#Show Distribution of Serum Cholestrol
df['chol'].hist()
```

```python
#plot Continuous Variables
cate_val=[]
cont_val=[]
```
